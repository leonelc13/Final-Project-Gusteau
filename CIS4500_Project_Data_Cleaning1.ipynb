```python
import pandas as pd

ingr_df = pd.read_csv('final_ingredients.csv')
prices_df = pd.read_csv('pricesFINAL.csv')
```

```python
prices_df['Ingredient_name'] = prices_df['Ingredient_name'].apply(lambda x: x.strip())
```

```python
prices_df.to_csv('final_prices.csv')
```

```python
ingr_set = sorted(set(ingr_df['Ingredient_name']))

def get_ingr_name(raw_ingr_name):
  good_match_exists = False
  potential_matches = [None] * 3 # 0 is perfect, 1 is matches prefix, 2 is contained
  for ingr in ingr_set:
    if ingr[:len(raw_ingr_name)] == raw_ingr_name:
      if len(ingr) == len(raw_ingr_name):
        potential_matches[0] = ingr
      else:
        potential_matches[1] = ingr
    elif ingr.find(raw_ingr_name) != -1:
      potential_matches[2] = ingr 
  
  for p in potential_matches:
    if p is not None:
      return p
  return ''

series = prices_df['Ingredient_name'].apply(get_ingr_name)
```

```python
prices_df['real_ingredient_name'] = series
```

```python
prices_df
```

```python
prices_df.to_csv('final_updated_prices.csv')
```
